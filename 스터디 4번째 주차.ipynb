```python
class TorchvisionDataset(Dataset):
    def __init__(self, file_paths, labels, transform=None):
        self.file_paths = file_paths
        self.labels = labels
        self.transform = transform
        
    def __len__(self):
        return len(self.file_paths)

    def __getitem__(self, idx):
        label = self.labels[idx]
        file_path = self.file_paths[idx]
        
        # Read an image with PIL
        image = Image.open(file_path)
        
        start_t = time.time()
        if self.transform:
            image = self.transform(image)
        total_time = (time.time() - start_t)

        return image, label, total_time
```

```python
class catdogDataset(Dataset): 
  def __init__(self, path, train=True, transform=None): 
    self.path = path 
    if train: 
      self.one_path = path + '/one' 
      self.two_path = path + '/two' 
    else: 
      self.cat_path = path + '/cat/test' 
      self.dog_path = path + '/dog/test' 
      
    self.one_img_list = glob.glob(self.one_path + '/*.jpg') 
    self.two_img_list = glob.glob(self.two_path + '/*.jpg') 
    self.transform = transform 
    self.img_list = self.one_img_list + self.two_img_list 
    self.class_list = [0] * len(self.one_img_list) + [1] * len(self.two_img_list) 
  
  def __len__(self): 
    return len(self.img_list) 
  
  def __getitem__(self, idx): 
    
    img_path = self.img_list[idx] 
    label = self.class_list[idx] 
    img = Image.open(img_path) 
    
    if self.transform is not None: 
      img = self.transform(img) 
    return img, label

```

```python
class Custom_Dataset(Dataset):
  def __setup_files(self):
    files = glob.glob(path+"/one"+"/*.jpg")
    return files
  
  def __init__(self,path,res,transforms=None):
    self.path = path
    self.res = res
    self.transforms = transforms
    self.files = self.__setup_files()


  def __len__(self):
    return len(__self.files)

  def __getitem__(self,idx):
    img = cv2.imread(self.files[idx])
    im_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    img = Image.fromarray(im_rgb.transpose(1,0,2))

    if self.transforms is not None:
      img = self.transforms(img)

    return img


```

```python
# for reproducibility
random.seed(777)
torch.manual_seed(777)
if device == 'cuda':
    torch.cuda.manual_seed_all(777)
```

```python
# hyperparameters
training_epochs = 15
batch_size = 100
```

```python
# MNIST dataset
mnist_train = dsets.MNIST(root='MNIST_data/',
                          train=True,
                          transform=transforms.ToTensor(),
                          download=True)

mnist_test = dsets.MNIST(root='MNIST_data/',
                         train=False,
                         transform=transforms.ToTensor(),
                         download=True)
```

```python
# dataset loader
data_loader = DataLoader(dataset=mnist_train,
                                          batch_size=batch_size, # 배치 크기는 100
                                          shuffle=True,
                                          drop_last=True)
```

```python
# 비용 함수와 옵티마이저 정의
criterion = nn.CrossEntropyLoss().to(device) # 내부적으로 소프트맥스 함수를 포함하고 있음.
optimizer = torch.optim.SGD(model_resnet18.parameters(), lr=0.1)
```

```python
for epoch in range(training_epochs): # 앞서 training_epochs의 값은 15로 지정함.
    avg_cost = 0
    total_batch = len(data_loader)

    for X, Y in data_loader:
        # 배치 크기가 100이므로 아래의 연산에서 X는 (100, 784)의 텐서가 된다.
        X = X.to(device)
        # 레이블은 원-핫 인코딩이 된 상태가 아니라 0 ~ 9의 정수.
        Y = Y.to(device)

        optimizer.zero_grad()
        hypothesis = model_resnet18(X)
        cost = criterion(hypothesis, Y)
        cost.backward()
        optimizer.step()

        avg_cost += cost / total_batch

    print('Epoch:', '%04d' % (epoch + 1), 'cost =', '{:.9f}'.format(avg_cost))

print('Learning finished')
```

```python
# 테스트 데이터를 사용하여 모델을 테스트한다.
with torch.no_grad(): # torch.no_grad()를 하면 gradient 계산을 수행하지 않는다.
    X_test = mnist_test.test_data.float().to(device)
    Y_test = mnist_test.test_labels.to(device)

    prediction = model_resnet18(X_test.view(-1,1,28,28))
    correct_prediction = torch.argmax(prediction, 1) == Y_test
    accuracy = correct_prediction.float().mean()
    print('Accuracy:', accuracy.item())
```
